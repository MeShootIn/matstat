```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def standard_normal(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x*x / 2)

def standard_cauchy(x):
    return 1 / (np.pi * (1 + x*x))

def laplace(x):
    return (1 / np.sqrt(2)) * np.exp(-np.sqrt(2) * np.abs(x))

def poisson(k):
    return (np.power(k, 10) / np.math.factorial(10)) * np.exp(-k)

def uniform(x):
    return (1 / (2 * np.sqrt(3))) * (x <= np.sqrt(3))

def laplace_gen(n):
    return np.random.laplace(0, 1 / np.sqrt(2), n)

def poisson_gen(n):
    return np.random.poisson(10, n)

def uniform_gen(n):
    return np.random.uniform(-np.sqrt(3), np.sqrt(3), n)

distr_func = {
    'standard_normal' : standard_normal,
    'standard_cauchy' : standard_cauchy,
    'laplace' : laplace,
    'poisson' : poisson,
    'uniform' : uniform
}

distr_gen = {
    'standard_normal' : np.random.standard_normal,
    'standard_cauchy' : np.random.standard_cauchy,
    'laplace' : laplace_gen,
    'poisson' : poisson_gen,
    'uniform' : uniform_gen
}

power = [10, 50, 100, 1000]
np.random.seed(0)
```

```python
for distr_name in distr_func.keys():
    for n in power:
        fig, ax = plt.subplots()
        plt.tight_layout()
        
        data = distr_gen[distr_name](n)
        plt.title(distr_name + str(' n=') + str(n))
        x = np.linspace(np.min(data), np.max(data), n)
        y = distr_func[distr_name](x)
        
        plt.hist(data, bins = 20, density = True)
        
        ax.plot(x, y, 'r')
        fig.savefig(distr_name + '_' + str(n))
```
